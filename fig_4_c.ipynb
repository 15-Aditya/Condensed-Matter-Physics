```python
import numpy as np
import matplotlib.pyplot as plt
import math

H = np.zeros((2,2), dtype = complex)
w = 0.7
u = 0.5
v = 0.8
eta_list = []
eta_der = []

H[0][0] = complex(0, u)
H[1][1] = complex(0, -u)

for k in np.linspace(-math.pi, math.pi, 201, endpoint = 'True'):
    #Initialising Hamiltonian
    H[0][1] = complex(w*math.cos(k)+v, -w*math.sin(k))
    H[1][0] = complex(w*math.cos(k)+v, w*math.sin(k))
    
    #Calculating eigenvalues and eigenvectors
    eigval, eigvec_arr = np.linalg.eig(H)
    eigvec = np.transpose(eigvec_arr)

    #Calculating eta
    eta = (np.absolute(np.inner(np.conjugate(eigvec[0]), eigvec[1])))**2
    eta_list.append(eta)

for m in range (0, 200):
    eta_der = (eta_list[m+1] - eta_list[m])/((2*math.pi)/200)
    #Plotting derivative of eta
    x_val = -math.pi + (m+1)*(2*math.pi)/200
    plt.plot(x_val, eta_der, marker = 'o', markersize = '2', color = 'blue')
    plt.xlabel('$\it{k}$')
    plt.ylabel('$\dfrac{d\eta}{dk}$')
    plt.xlim(-math.pi-0.1, math.pi+0.1)
    plt.ylim(-3, 3)

plt.show()





```
