```python
import numpy as np
import matplotlib.pyplot as plt
import math

H = np.zeros((2,2), dtype = complex)
w = 0.7
u = 0.5
v = 0.8

H[0][0] = complex(0, u)
H[1][1] = complex(0, -u)

for k in np.linspace(-math.pi, math.pi, 200, endpoint = 'True'):
    H[0][1] = complex(w*math.cos(k)+v, -w*math.sin(k))
    H[1][0] = complex(w*math.cos(k)+v, w*math.sin(k))
    
    eigval, eigvec_arr = np.linalg.eig(H)
    eigvec = np.transpose(eigvec_arr)

    eta = (np.absolute(np.inner(np.conjugate(eigvec[0]), eigvec[1])))**2
    plt.plot(k, eta, marker = 'o', markersize = '2', color = 'blue')
    plt.xlabel('$\it{k}$')
    plt.ylabel('$\eta$')
    plt.xlim(-math.pi-0.01, math.pi+0.01)
plt.show()





```
