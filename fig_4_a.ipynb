```python
#The plot given in research paper has wrong index, just put k = 0 and verify, for k = 0 we get real eigenvalues
import numpy as np
import matplotlib.pyplot as plt
import math

#Initialising variables
H = np.zeros((2,2), dtype = complex)
w = 0.7 
u = 0.5
v = 0.8 

#Initialising diagonal elements of Hamilotnian
H[0][0] = complex(0,u)
H[1][1] = complex(0,-u)

for k in np.linspace(-math.pi, math.pi, 101, endpoint = 'True'):

    #Initialising off-diagonal elements
    H[0][1] = w*complex(math.cos(k), -math.sin(k)) + v
    H[1][0] = w*complex(math.cos(k), math.sin(k)) + v

    #Calculating eigenvalues
    eigval = np.linalg.eigvals(H)

    #Plotting real part of eigenvalues
    plt.plot(k, np.real(eigval[0]), marker = 'o', markersize = '2', color = 'red')
    plt.plot(k, np.real(eigval[1]), marker = 'o', markersize = '2', color = 'red')

    #Plotting imaginary part of eigenvalues
    plt.plot(k, np.imag(eigval[0]), marker = 'o', markersize = '2', color = 'blue')
    plt.plot(k, np.imag(eigval[1]), marker = 'o', markersize = '2', color = 'blue')

    plt.xlabel('$\it{k}$')
    plt.ylabel('$E_{k}$')
    plt.ylim(-2, 2)

plt.show()
```
