Energy Spectrum SSH Model N = 4

```python
import numpy as np
import matplotlib.pyplot as plt

t_2 = 2
N = 4

H = np.zeros((N, N))

for t_1 in np.linspace(0, 5, 101, endpoint = True):
    for i in range (0, N-1):
        if(i % 2 == 0):
            H[i,i+1] = -t_1
            H[i+1,i] = -t_1
        if(i % 2 == 1 and i < N):
            H[i,i+1] = -t_2
            H[i+1,i] = -t_2
        eigval, eigvec = np.linalg.eigh(H)
        for j in range (0, N):
            plt.plot(t_1, eigval[j], marker = 'o', markersize = '2', color = 'blue')
            plt.xticks(np.arange(0, N+2, 1))
            plt.xlabel("$t_{1}$")
            plt.ylabel('E')
            plt.xlim(0,5.5)

```

Metallic Phase SSH Model N = 4
$$t_{1} = t_{2}$$

```python
import numpy as np
import matplotlib.pyplot as plt

t_1 = 2
t_2 = 2
N = 4

H = np.zeros((N, N))

for i in range (0, N-1):
    if(i % 2 == 0):
        H[i,i+1] = -t_1
        H[i+1,i] = -t_1
    if(i % 2 == 1 and i < N):
        H[i,i+1] = -t_2
        H[i+1,i] = -t_2

print(H)
eigval, eigvec = np.linalg.eigh(H)
print(eigvec)

for i in range (0, N):
    prob_den = eigvec[i][2]**2      #probability density at nth site
    plt.bar(i+1, prob_den, width = 0.2, color = 'red')
    plt.xticks(np.arange(1, N+1, 1))
    plt.xlabel("Site")
    plt.ylabel('$|\psi|^{2}$')
```

Energy Spectrum SSH Model with on-site potential 

```python
import numpy as np
import matplotlib.pyplot as plt

t_2 = 2
mu_1 = 1
mu_2 = 1
mu_3 = 1.3
mu_4 = 1.5

N = 4

H = np.zeros((N, N))

H[0,0] = mu_1
H[1,1] = mu_2
H[2,2] = mu_3
H[3,3] = mu_4

for t_1 in np.linspace(0, 5, 101, endpoint = True):
    for i in range (0, N-1):
        if(i % 2 == 0):
            H[i,i+1] = -t_1
            H[i+1,i] = -t_1
        if(i % 2 == 1 and i < N):
            H[i,i+1] = -t_2
            H[i+1,i] = -t_2
        eigval, eigvec = np.linalg.eigh(H)
        for j in range (0, N):
            plt.plot(t_1, eigval[j], marker = 'o', markersize = '2', color = 'blue')
            plt.xticks(np.arange(0, N+2, 1))
            plt.xlabel("$t_{1}$")
            plt.ylabel('E')
            plt.xlim(0, 5.5)

```
