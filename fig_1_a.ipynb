```python
import matplotlib.pyplot as mpl
import numpy

#Initialising variables
N = 150
eta = []
H = numpy.zeros((N,N))

for t_1 in numpy.linspace(0.0, 2.0, num = 41, endpoint = True):
    
    #Initialising variables 
    eigval = []

    #Generating Hamiltonian
    for i in range (0, N-1):
        if(i % 2 == 0):
            H[i,i+1] = t_1 #t_1
            H[i+1,i] = t_1 #t_1
        elif(i % 2 == 1 and i < N-1):
            H[i,i+1] = 1.1 #t_2+g
            H[i+1,i] = 0.9 #t_2-g
    
    #Calculating eigenvalues and eigenvectors
    eigval = numpy.linalg.eigvals(H)
    real_eigval = numpy.linalg.eigvals(H).real

    #Calculating eta
    for i in range (0,N):
        mpl.plot(t_1, real_eigval[i], 'o', markersize = '3', color = 'black')
mpl.show()
```
