```python
import matplotlib.pyplot as mpl
import numpy

num = 0.0
den = 0.0
eta_der = []

for gamma in numpy.arange(-1.51, -0.05, 0.05):
    num = 4*(gamma+1)
    den = (1-gamma)**3
    eta_der.append(num / den)

for gamma in numpy.arange(0, 1.51, 0.05):
    num = 4*(gamma-1)
    den = (1+gamma)**3
    eta_der.append(num / den)

gamma = numpy.arange(-1.5, 1.51, 0.05)
mpl.plot(gamma, eta_der, 'o', markersize = '3', color = 'black')
mpl.show()
```
