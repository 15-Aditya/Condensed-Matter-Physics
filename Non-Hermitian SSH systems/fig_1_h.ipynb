```python
import matplotlib.pyplot as mpl
import numpy

H = numpy.zeros((120,120))
for i in range (0, 119):
    if(i % 2 == 0):
        H[i,i+1] = -0.22 #t_1
        H[i+1,i] = -0.44 #t_1/g, g = 0.5
    elif(i % 2 == 1 and i < 119):
        H[i,i+1] = 2.0 #t_2/g, g = 0.5
        H[i+1,i] = 1.0 #t_2
eigval, eigvec = numpy.linalg.eig(H)
eig_vec = numpy.transpose(eigvec)
for j in range (0, 120):
    mpl.plot(j, (eig_vec[0][j])*(eig_vec[0][j]), marker = 'o', markersize = '2', color = 'blue')
    mpl.plot(j, (eig_vec[119][j])*(eig_vec[119][j]), marker = 'o', markersize = '2', color = 'orange')
mpl.ylim(-0.1,1)
mpl.show()    
```
