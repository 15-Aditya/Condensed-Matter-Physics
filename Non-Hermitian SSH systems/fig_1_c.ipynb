```python
import matplotlib.pyplot as mpl
import numpy

N = 120
eta = []
H = numpy.zeros((N,N))
for t_1 in numpy.linspace(-1.0, 1.0, num = 41, endpoint = True):
    num = 0.0
    den = 0.0
    eigval = []
    abs_eigval = []
    for i in range (0, N-1):
        if(i % 2 == 0):
            H[i,i+1] = t_1 #t_1
            H[i+1,i] = t_1 #t_1/g, g = 1
        elif(i % 2 == 1 and i < N-1):
            H[i,i+1] = 1.0 #t_2/g
            H[i+1,i] = 1.0 #t_2
    eigval = numpy.linalg.eigvals(H)
    abs_eigval = abs(numpy.linalg.eigvals(H))
    for i in range (0,N):
        mpl.plot(t_1, abs_eigval[i], 'o', markersize = '3', color = 'black')
        mpl.plot(t_1, -abs_eigval[i], 'o', markersize = '3', color = 'black')
mpl.show()
```
