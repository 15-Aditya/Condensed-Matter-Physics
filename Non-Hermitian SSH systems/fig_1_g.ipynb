```python
import matplotlib.pyplot as mpl
import numpy

H = numpy.zeros((120,120))
for i in range (0, 119):
    if(i % 2 == 0):
        H[i,i+1] = 0 #t_1
        H[i+1,i] = 0 #t_1
    elif(i % 2 == 1 and i < 119):
        H[i,i+1] = 2 #t_2, g = 0.5
        H[i+1,i] = 1 #t_2
eigval, eigvec = numpy.linalg.eig(H)
for j in range (0, 120):
    mpl.plot(j, eigvec[j][118], marker = 'o', markersize = '2', color = 'blue')
    mpl.plot(j, eigvec[j][119], marker = 'o', markersize = '2', color = 'orange')
mpl.show()    
```
